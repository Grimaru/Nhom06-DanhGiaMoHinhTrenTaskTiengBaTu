```python
from evaluate import load

evaluator = load("competition_math")
```

```python
def preserve_digit(input_string):
    end_idx = len(input_string) - 1
    while end_idx >= 0 and not input_string[end_idx].isdigit():
        end_idx -= 1
    start_idx = end_idx

    while (input_string[start_idx].isdigit() or input_string[start_idx] in ['.', ',', "/", "\\", "{", "}"]) and start_idx > 0:
        start_idx -= 1

    return input_string[start_idx+1:end_idx+1]
```

```python
sample_GPT_response = """\
[راه حل]

مثلث اول با اضلاع 6، 8 و 10 به ترتیب، مثلث متساوی الساقین است.

ابتدا مساحت مثلث اول را محاسبه می‌کنیم. از فرمول هرون استفاده می‌کنیم:

\(s = \frac{a + b + c}{2}\)

\(s = \frac{6 + 8 + 10}{2} = 12\)

حالا از فرمول مساحت هرون:

\(Area = \sqrt{s \cdot (s - a) \cdot (s - b) \cdot (s - c)}\)

\(Area = \sqrt{12 \cdot (12 - 6) \cdot (12 - 8) \cdot (12 - 10)}\)

\(Area = \sqrt{12 \cdot 6 \cdot 4 \cdot 2} = \sqrt{576} = 24\)

مساحت مثلث اول برابر با 24 واحد مربع است.

حالا طول هر یک از اضلاع را 50% افزایش می‌دهیم:

طول جدید \(a\) = \(6 + 0.5 \times 6 = 9\)

طول جدید \(b\) = \(8 + 0.5 \times 8 = 12\)

طول جدید \(c\) = \(10 + 0.5 \times 10 = 15\)

سپس مساحت مثلث جدید را محاسبه می‌کنیم:

\(s_{new} = \frac{a_{new} + b_{new} + c_{new}}{2} = \frac{9 + 12 + 15}{2} = 18\)

\(Area_{new} = \sqrt{s_{new} \cdot (s_{new} - a_{new}) \cdot (s_{new} - b_{new}) \cdot (s_{new} - c_{new})}\)

\(Area_{new} = \sqrt{18 \cdot (18 - 9) \cdot (18 - 12) \cdot (18 - 15)}\)

\(Area_{new} = \sqrt{18 \cdot 9 \cdot 6 \cdot 3} = \sqrt{2916} = 54\)

پس مساحت مثلث جدید برابر با 54 واحد مربع می‌باشد.

[پاسخ]

54
"""
```

```python
sample_result_idx = sample_GPT_response.rfind("[پاسخ]")
```

```python
sample_result = sample_GPT_response[sample_result_idx:]
```

```python
pred_labels = [preserve_digit(sample_result)]
true_labels = [sample['target']]
```

```python
results = evaluator.compute(references=true_labels, predictions=pred_labels)
```
