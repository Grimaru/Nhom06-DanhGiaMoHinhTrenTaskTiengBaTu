```python
from transformers import AutoTokenizer, AutoModelForCausalLM
import torch

# Tải mô hình bloomz-560m (có thể dùng bloomz-1b1, bloomz-7b1 nếu máy đủ mạnh)
tokenizer = AutoTokenizer.from_pretrained("cmarkea/bloomz-560m-sft-chat")
model = AutoModelForCausalLM.from_pretrained("cmarkea/bloomz-560m-sft-chat")

def get_completion(prompt):
    inputs = tokenizer(prompt, return_tensors="pt")
    outputs = model.generate(
        **inputs,
        max_new_tokens=100,
        do_sample=False,  # giống như temperature=0
    )
    return tokenizer.decode(outputs[0], skip_special_tokens=True)
```

```python
import time
results = {"question":[],"correct_ans":[],"model_ans":[]}

for i in range(len(df)):

  question = df['question'].values[i]
  correct_ans =  df['true_answer'].values[i]
  print(f'question {i+1}')

  prompt = f"""
  In this task, you will be shown a Persian elementary math question. You need to write an answer for the question.\
  the question is a two choice question.
  The answer is either `ا` or `ب`.\
  So you should answer only in one letter, `ا` or `ب`.

  question:
  '''{question}

  answer:
  """

  response = get_completion(prompt)
  print(prompt)
  print("model ans: ",response)
  print("correct ans: ",correct_ans)
  print("-------------------------------------------------------------------------------------")
  results['question'].append(question)
  results['correct_ans'].append(correct_ans)
  results['model_ans'].append(response)
  time.sleep(22)
```

```python
df_res = pd.DataFrame(results)
df_res.to_csv('elem_q_res (bloomz).csv')
```

```python
df_i = pd.read_csv('elem_q_res (bloomz).csv')
gpt_results = df_i['model_ans'].values
true_ans = [str(x) for x in df["true_answer"].values]
acc = (gpt_results == true_ans).sum()/len(true_ans)
acc
```

## one shot

```python
import random
results = {"question":[],"correct_ans":[],"model_ans_one_shot":[]}

for i in range(len(df)):
  idx = random.randint(0, len(df))
  while True:
      if idx == i:
          idx = random.randint(0, len(df))
      else:
          break
  question = df['question'].values[i]
  correct_ans =  df['true_answer'].values[i]
  print(f'question {i+1}')

  prompt = f"""
  In this task, you will be shown a Persian elementary math question. You need to write an answer for the question.\
  the question is a two choice question.
  The answer is either `ا` or `ب`.\
  So you should answer only in one letter, `ا` or `ب`.
  here is an example:
  question:
    {df.iloc[idx]["question"]}
  answer:
    {df.iloc[idx]["true_answer"]}
  question:
  '''{question}

  answer:
  """

  response = get_completion(prompt)
  print(prompt)
  print("model ans: ",response)
  print("correct ans: ",correct_ans)
  print("-------------------------------------------------------------------------------------")
  results['question'].append(question)
  results['correct_ans'].append(correct_ans)
  results['model_ans_one_shot'].append(response)
  time.sleep(21)
```
