```python
from openai import OpenAI

# Khởi tạo client
client = OpenAI(api_key="--")

def get_completion(prompt, model="gpt-3.5-turbo"):
    response = client.chat.completions.create(
        model=model,
        messages=[{"role": "user", "content": prompt}],
        temperature=0
    )
    return response.choices[0].message.content.strip()
```

```python
results = {"question":[],"correct_ans":[],"model_ans":[]}

for i in range(len(df)):

  question = df['question'].values[i]
  correct_ans =  df['true_answer'].values[i]
  print(f'question {i+1}')

  prompt = f"""
  In this task, you will be shown a Persian elementary math question. You need to write an answer for the question.\
  the question is a two choice question.
  The answer is either `ا` or `ب`.\
  So you should answer only in one letter, `ا` or `ب`.

  question:
  '''{question}

  answer:
  """

  response = get_completion(prompt)
  print(prompt)
  print("model ans: ",response)
  print("correct ans: ",correct_ans)
  print("-------------------------------------------------------------------------------------")
  results['question'].append(question)
  results['correct_ans'].append(correct_ans)
  results['model_ans'].append(response)
  time.sleep(22)
```

## one shot

```python
results = {"question":[],"correct_ans":[],"model_ans_one_shot":[]}

for i in range(len(df)):
  idx = random.randint(0, len(df))
  while True:
      if idx == i:
          idx = random.randint(0, len(df))
      else:
          break
  question = df['question'].values[i]
  correct_ans =  df['true_answer'].values[i]
  print(f'question {i+1}')

  prompt = f"""
  In this task, you will be shown a Persian elementary math question. You need to write an answer for the question.\
  the question is a two choice question.
  The answer is either `ا` or `ب`.\
  So you should answer only in one letter, `ا` or `ب`.
  here is an example:
  question:
    {df.iloc[idx]["question"]}
  answer:
    {df.iloc[idx]["true_answer"]}
  question:
  '''{question}

  answer:
  """

  response = get_completion(prompt)
  print(prompt)
  print("model ans: ",response)
  print("correct ans: ",correct_ans)
  print("-------------------------------------------------------------------------------------")
  results['question'].append(question)
  results['correct_ans'].append(correct_ans)
  results['model_ans_one_shot'].append(response)
  time.sleep(21)
```

# Evaluate

```python
import pandas as pd
df = pd.read_csv("elem_q_res.csv")
```

```python
answer_list_pp = {}
accuracy = {}
a = list()
for i in range(len(df)):
  a.append(0)
for i in range(len(df.columns[3:])):
  answer_list_pp[df.columns[i+3]] = a
```

```python
for i in range(len(df.columns[3:])):
  c = 0

  for j in range(len(df)):
    answer_list_pp[df.columns[i+3]][j] = df[df.columns[i+3]][j]
    if df["correct_ans"][j] in df[df.columns[i+3]][j][:1]:
      answer_list_pp[df.columns[i+3]][j] = df.correct_ans[j]
      c += 1
    else:
      answer_list_pp[df.columns[i+3]][j] = answer_list_pp[df.columns[i+3]][j][:1]
  accuracy[df.columns[i+3]] = c/len(df)
```

```python
for i,j in enumerate(answer_list_pp.keys()):
  df[df.columns[i+3] + "_post_process"] = answer_list_pp[j]
```

```python
df.to_csv("result_elem_q_pp.csv")
```
